### Building Popularity Model

```python
#Count of user_id for each unique product as recommendation score 
train_data_grouped = train_data.groupby('productId').agg({'useId': 'count'}).reset_index()
train_data_grouped.rename(columns = {'useId': 'score'},inplace=True)
train_data_grouped.head(40)
```

```python
#Sort the products on recommendation score 
train_data_sort=train_data_grouped.sort_values(['score','productId'],ascending=False) # ascending=[0,1]

#Generate a recommendation rank based upon score 
train_data_sort['rank']=train_data_sort['score'].rank(ascending=False,method='first')
#train_data_sort

#Get the top 10 recommendations 
popularity_recommendation=train_data_sort.head(10)
popularity_recommendation
```

```python
def recommend(user_id):     
    user_recommendations = popularity_recommendation 
          
    #Add user_id column for which the recommendations are being generated 
    user_recommendations['useId'] = user_id 
      
    #Bring user_id column to the front 
    cols = user_recommendations.columns.tolist() 
    cols = cols[-1:] + cols[:-1] 
    user_recommendations = user_recommendations[cols] 
          
    return user_recommendations
```

```python
find_recom = [10,100,150]   # This list is user choice.
for i in find_recom:
    print("The list of recommendations for the userId: %d\n" %(i))
    print(recommend(i))    
    print("\n")
```

### Collaborative Filtering Recommender Model

```python
df_cf= pd.concat([train_data,test_data]).reset_index()
df_cf.head()
```

```python
df_cf.shape
```

### User Based Collaborative Filtering model

```python
# Matrix with row per 'user' and column per 'item' 
pivot_df_cf= df_cf.pivot(index='useId',columns='productId',values='ratings').fillna(0)
pivot_df_cf.head()
```

```python
pivot_df_cf.shape
```

```python
#define user index from 0 to 10
pivot_df_cf['user_index'] = np.arange(0,pivot_df_cf.shape[0], 1)
pivot_df_cf.head()
```

```python
pivot_df_cf.set_index(['user_index'],inplace=True)
pivot_df_cf.head()
```

### Singular Value Decomposition

```python
U,sigma,Vt=svds(pivot_df_cf,k=10)
print('Left singular matrix: \n',U)
```

```python
print('sigma:\n',sigma)
```

```python
# Construct diagonal array in SVD
sigma = np.diag(sigma)
print('Diagonal matrix: \n',sigma)
```

```python
print('Right singular matrix: \n',Vt)
```

```python
#predicted ratings
all_user_predicted_rating=U.dot(sigma.dot(Vt))
preds_df=pd.DataFrame(all_user_predicted_rating,columns=pivot_df_cf.columns)
preds_df.head()
```

```python
# Recommend the items with the highest predicted ratings

def recommend_items(useID, pivot_df_cf, preds_df, num_recommendations):
    # index starts at 0  
    user_idx = useID-1 
    # Get and sort the user's ratings
    sorted_user_ratings = pivot_df_cf.iloc[user_idx].sort_values(ascending=False)
    #sorted_user_ratings
    sorted_user_predictions = preds_df.iloc[user_idx].sort_values(ascending=False)
    #sorted_user_predictions
    temp = pd.concat([sorted_user_ratings, sorted_user_predictions], axis=1)
    temp.index.name = 'Recommended Items'
    temp.columns = ['user_ratings', 'user_predictions']
    temp = temp.loc[temp.user_ratings == 0]   
    temp = temp.sort_values('user_predictions', ascending=False)
    print('\nBelow are the recommended items for user(user_id = {}):\n'.format(useID))
    print(temp.head(num_recommendations))
```

```python
useID = 4
num_recommendations = 5
recommend_items(useID, pivot_df_cf, preds_df, num_recommendations)
```
